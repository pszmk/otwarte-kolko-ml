### Macierze

##### Definicja

Macierz to tablica dwuwymiarowa zawierająca liczby, wyrażenia lub symbole.
(definicja z https://pl.wikipedia.org/wiki/Macierz)

##### Mnożenie macierzy

Macierze mnożymy w następujący sposób:



Każdy element wiersza pierwszej macierzy, mnożymy z odpowiadającym im elementą z kolumny drugiej macierzy, po czym dodajemy je do siebie.

Uwaga!
Dwóch macierzy nie możemy ponożyć, jeśli liczba wierszy jednej macierzy nie jest równa liczbie kolumn drugiej, lub liczba elementów w wierszach jest inna niż liczba elementów w kolumnach drugiej macierzy.



##### Przykład kodu

```python
# Implementacja pierwszej macierzy:
M1 = np.array([[w11, w12, w13],
              [w21, w22, w23],
              [w31, w32, w33]])
# Implementacja drugiej macierzy:
M2 = np.array([[k11, k12, k13],
              [k21, k22, k23],
              [k31, k32, k33]])
# Wyliczenie macierzy powstałej z mnożenia macierzy M1 i M2:
# copilot to dał, nie wiem czy działa
M3 = np.dot(M1, M2)
print(f"Macierz M3: {M3}")
```
